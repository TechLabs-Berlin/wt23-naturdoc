```python
from pytrends.request import TrendReq
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**PyTrends**

```python
# Set up the PyTrends client
pytrends = TrendReq(hl='de-DE', tz=360)

# Define the search query for herbs in Germany
#kw_list = ['Heilkräuter','Herbs','Natural Remedies','Naturheilmittel']
kw_list = ['Heilkräuter']

# Build the PyTrends payload
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='DE', gprop='')

# Print the top 5 most popular herbs searched for in Germany
df_over_time = pytrends.interest_over_time()
df_over_time.head(5)
```

```python
related_topics_dict = pytrends.related_topics()
related_topics_df = pd.DataFrame(related_topics_dict[kw_list[0]]["rising"])
related_topics_df.head(10)
```

```python
herbs_split = [herbs[i:i+5] for i in range(0, len(herbs), 5)]
herbs_split
```

```python

# Initialize Pytrends
pytrends = TrendReq()

# Create an empty list to store the search results for each herb
results = []

# Loop over each sublist of 5 herbs
for herb_group in herbs_split:
    # Convert each herb to a string and join them with commas
    herb_group_str = ",".join(str(herb) for herb in herb_group)
    
    # Build the Pytrends payload
    pytrends.build_payload(kw_list=[herb_group_str], timeframe='today 5-y', geo='DE')
    
    # Get the interest over time data for the herb
    interest_over_time_df = pytrends.interest_over_time()
    
if not interest_over_time_df.empty:
    total_interest = interest_over_time_df.sum().values[0]
else:
    total_interest = 0

    # Sum the interest scores for the herb over the last 5 years
    total_interest = interest_over_time_df.sum().values[0]
    
    # Add the herb and its total interest score to the results list
    results.append((herb_group_str, total_interest))

# Sort the results in descending order
results_sorted = sorted(results, key=lambda x: x[1], reverse=True)

# Print the top 10 most popular herbs
for i in range(10):
    print(results_sorted[i])
```
